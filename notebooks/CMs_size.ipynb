```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Lire sans noms de colonnes pour voir ce qu'il y a vraiment
raw = pd.read_csv(
    '/home/boussif/boussif/CM_inference/data/processed/SEA-AD-Neu/1.clomics_CMs/snATAC/n_peaks_200/bg_threshold_3/SEA-AD-Neu/SEA-AD-Neu_Clomics_CM.tracks.bed',
    sep='\t', header=None
)
print(f"Nombre de colonnes : {raw.shape[1]}")
print(raw.head(3).to_string())
```

```python
raw.columns = ['chr', 'start', 'end', 'cm_id', 'score', 'strand',
               'thick_start', 'thick_end', 'rgb', 'block_count',
               'block_sizes', 'block_starts', 'extra']

sizes = raw['end'] - raw['start']
print(f"Nombre de CMs : {len(raw)}")
print(f"\nTaille des CMs (kb) :")
print((sizes / 1000).describe())

plt.figure(figsize=(8, 4))
plt.hist(sizes / 1000, bins=1000, edgecolor='black')
plt.xlabel('Taille du CM (kb)')
plt.xlim(0, 1000)  # Limiter l'axe x pour mieux voir la distribution des petites tailles
plt.ylabel('Nombre de CMs')
plt.title(f'Distribution des tailles (médiane = {sizes.median()/1000:.1f} kb)')
plt.axvline(sizes.median() / 1000, color='red', linestyle='--', label='Médiane')
plt.legend()
plt.tight_layout()
plt.savefig('cm_sizes.png', dpi=150)
plt.show()
```
